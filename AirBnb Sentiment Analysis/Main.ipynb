```python
# Install necessary libraries if not installed
!pip install langdetect swifter

# Import required libraries
import pandas as pd
import swifter
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

# Fix randomness in language detection
DetectorFactory.seed = 42

# Remove duplicate reviews if any exist
df_merged.drop_duplicates(inplace=True)

# Convert the date column to datetime format
df_merged['date'] = pd.to_datetime(df_merged['date'], errors='coerce')

# Keep only reviews from 2015 onwards
df_merged = df_merged[df_merged['date'].dt.year >= 2015]

# Randomly sample a subset of data (adjust sample size if needed)
df_sampled = df_merged.sample(n=200000, random_state=42)

# Function to detect language (optimized: only use first 200 characters)
def detect_language(text):
    try:
        return detect(text[:200])  # Check only the first 200 characters
    except LangDetectException:
        return "unknown"  # If detection fails, mark as unknown

# Apply language detection using swifter for parallel processing
df_sampled['language'] = df_sampled['comments'].swifter.apply(detect_language)

# Keep only English comments
df_sampled = df_sampled[df_sampled['language'] == 'en']

# Drop the language column after filtering
df_sampled.drop(columns=['language'], inplace=True)

# Display final dataset shape
print("\nFinal Dataset Shape After Cleaning and Language Filtering:", df_sampled.shape)

# Show a sample of the final dataset
print("\nSample of English-Only Dataset:")
display(df_sampled.head())
```

```python
# Import necessary libraries
import spacy
import pandas as pd

# Load the English NLP model with unnecessary components disabled to speed up processing
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser", "textcat"])

# Function to process text in batches
def clean_text_spacy(texts):
    """
    Cleans text by:
    - Lowercasing
    - Removing punctuation and special characters
    - Removing stopwords
    - Lemmatizing words
    """
    cleaned_texts = []
    
    # Process text in batches for efficiency
    for doc in nlp.pipe(texts, batch_size=500):
        tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]  # Keep only alphabetic words
        cleaned_texts.append(" ".join(tokens))  # Join cleaned tokens back into text
    
    return cleaned_texts

# Ensure the comments are strings and apply cleaning function
df_sampled['cleaned_comments'] = clean_text_spacy(df_sampled['comments'].astype(str))

# Display sample of cleaned comments
print("Cleaned Comments Sample:")
display(df_sampled[['comments', 'cleaned_comments']].head())

```

```python
# Install required libraries
!pip install scikit-learn

# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define positive and negative descriptors
positive_words = ["vibrant", "charming", "peaceful", "gastronomic", "affordable", "historic"]
negative_words = ["isolated", "overpriced", "noisy", "unsafe", "crime", "gentrified"]

# Convert descriptor lists into text format
positive_text = " ".join(positive_words)
negative_text = " ".join(negative_words)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_sampled['cleaned_comments'])  # Transform reviews
descriptor_matrix = vectorizer.transform([positive_text, negative_text])  # Transform descriptors

# Compute cosine similarity between each review and descriptors
positive_similarity = cosine_similarity(tfidf_matrix, descriptor_matrix[0])  # Similarity with positive words
negative_similarity = cosine_similarity(tfidf_matrix, descriptor_matrix[1])  # Similarity with negative words

# Convert similarity scores into a DataFrame
df_sampled['positive_similarity'] = positive_similarity.flatten()
df_sampled['negative_similarity'] = negative_similarity.flatten()

# Aggregate similarity scores at the neighborhood level
df_neighborhood_similarity = df_sampled.groupby('neighbourhood_cleansed').agg({
    'positive_similarity': 'mean',
    'negative_similarity': 'mean'
}).reset_index()

# Display results
print("\nNeighborhood Similarity Scores:")
display(df_neighborhood_similarity.head())
```

```python
# Install necessary libraries if not installed
!pip install matplotlib seaborn

# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Select only numerical columns for correlation
numeric_cols = ['positive_similarity', 'negative_similarity']
df_corr = df_neighborhood_similarity[numeric_cols].corr()

# Create a scatter plot to show correlation between positive and negative similarity
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_neighborhood_similarity, 
    x='positive_similarity', 
    y='negative_similarity', 
    hue='neighbourhood_cleansed', 
    palette='viridis', 
    alpha=0.7
)
plt.xlabel("Positive Descriptor Similarity")
plt.ylabel("Negative Descriptor Similarity")
plt.title("Neighborhood Similarity Correlation")
plt.legend([],[], frameon=False)  # Hide legend if too many neighborhoods
plt.show()

# Create a heatmap of correlation
plt.figure(figsize=(8, 5))
sns.heatmap(df_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Positive & Negative Similarity Scores")
plt.show()

# Create a bar chart ranking neighborhoods by positive similarity
df_neighborhood_similarity_sorted = df_neighborhood_similarity.sort_values(by='positive_similarity', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_neighborhood_similarity_sorted.head(15), x='positive_similarity', y='neighbourhood_cleansed', palette="Blues_r")
plt.xlabel("Positive Similarity Score")
plt.ylabel("Neighborhood")
plt.title("Top 15 Most Positively Similar Neighborhoods")
plt.show()

# Create a bar chart ranking neighborhoods by negative similarity
df_neighborhood_similarity_sorted = df_neighborhood_similarity.sort_values(by='negative_similarity', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_neighborhood_similarity_sorted.head(15), x='negative_similarity', y='neighbourhood_cleansed', palette="Reds_r")
plt.xlabel("Negative Similarity Score")
plt.ylabel("Neighborhood")
plt.title("Top 15 Most Negatively Similar Neighborhoods")
plt.show()
```

```python
# Install necessary library if not installed
!pip install vaderSentiment

# Import required libraries
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to compute sentiment scores
def get_sentiment(text):
    sentiment = analyzer.polarity_scores(text)
    return sentiment['compound']  # Compound score summarizes overall sentiment

# Apply sentiment analysis to the cleaned comments
df_sampled['sentiment_score'] = df_sampled['cleaned_comments'].apply(get_sentiment)

# Aggregate sentiment scores by neighborhood
df_neighborhood_sentiment = df_sampled.groupby('neighbourhood_cleansed').agg({
    'sentiment_score': 'mean'
}).reset_index()

# Sort neighborhoods by sentiment
df_neighborhood_sentiment_sorted = df_neighborhood_sentiment.sort_values(by='sentiment_score', ascending=False)

# Display top 10 happiest neighborhoods
print("\nTop 10 Happiest Neighborhoods (Highest Sentiment Scores):")
print(df_neighborhood_sentiment_sorted.head(10).to_string(index=False))

# Display top 10 most negatively reviewed neighborhoods
print("\nTop 10 Most Negatively Reviewed Neighborhoods (Lowest Sentiment Scores):")
print(df_neighborhood_sentiment_sorted.tail(10).to_string(index=False))
```

```python
# Ensure date is in datetime format
df_sampled['date'] = pd.to_datetime(df_sampled['date'], errors='coerce')

# Extract year from the date
df_sampled['year'] = df_sampled['date'].dt.year

# Aggregate sentiment scores by year
df_yearly_sentiment = df_sampled.groupby('year').agg({
    'sentiment_score': 'mean'
}).reset_index()

# Display sentiment scores per year
print("\nSentiment Trends Over the Years:")
print(df_yearly_sentiment.to_string(index=False))

# Calculate variance to check stability
sentiment_variance = df_yearly_sentiment['sentiment_score'].var()
print(f"\nVariance of Sentiment Scores Over the Years: {sentiment_variance:.4f}")

# Interpret Stability
if sentiment_variance < 0.01:
    print("\nThe sentiment trends are STABLE over the years.")
else:
    print("\nThe sentiment trends are FLUCTUATING over the years.")
```
